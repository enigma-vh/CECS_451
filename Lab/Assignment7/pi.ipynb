```python
import matplotlib.pyplot as plt
import random
import numpy as np
import math
```

```python
def distance(x, y):
    return math.sqrt(x**2 + y**2)


def result(n, nums_in):
    estimated_Pi = 4 * (nums_in/n)
    errorRate = abs((math.pi - estimated_Pi)/math.pi) * 100
    return estimated_Pi, errorRate


def simulation(n):
    p_in = []
    p_out = []
    for i in range(n):
        x = random.random()
        y = random.random()
        if distance(x, y) < 1:
            p_in.append((x,y))
        else:
            p_out.append((x,y))
    return p_in, p_out
```

```python
def plot(p_in, p_out):
    plt.figure(figsize=(8,8))
    zip_in = zip(*p_in)
    plt.scatter(*zip_in, s = 0.5, color='b')
    zip_out = zip(*p_out)
    plt.scatter(*zip_out, s = 0.5, color='r')
    plt.axis('square')
    plt
```
